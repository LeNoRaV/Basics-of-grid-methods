```python
import pandas as pd
import numpy as np
import scipy as sp
import math
import matplotlib.pyplot as plt
import seaborn as sb
from prettytable import PrettyTable

np.set_printoptions(precision=3, suppress=True)
```

```python
N=8
n=62
a=0
b=(N+3.0)/N
lambd=1/(n-59)
knot=40
```

```python
h=(b-a)/knot
A=[]
sum=0
for i in range (knot):
  A.append(sum)
  sum+=h
A.append(sum)

print("a =", a)
print("b =", b)
print("lambda =", lambd)
print("knot =", knot)
print("A =", A)

B=[]
for i in range (1, knot+1):
  B.append((A[i-1]+A[i])/2)
print("B =", B)
```

```python
y=[]
for s in B:
  y.append((N+3)/N*(2*s*s*s-s*s+n-67))
print("y =", y)
```
