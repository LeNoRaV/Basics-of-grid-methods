```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
h=0.05
tau=h
betta=4
alpha=0.5
k=20
print(k)
```

```python
A=np.arange(0, 1+tau, tau)
B=np.arange(0, 1+h, h)
C=np.column_stack([A, B])
print(C)
```

```python
D=1
p=D*tau/h**2
q=1-2*D*tau/h**2
r=D*tau/h**2
G=np.zeros((21, 21))
for i in range(1,k):
  G[i][i-1]=p
  G[i][i]=q
  G[i][i+1]=r
print(pd.DataFrame(G))
```

```python
b=np.zeros((k+1,k+1))
i=0
for t in A:
  b[i][0]=2*betta*(1-t)
  b[i][k]=2*betta*(1-t)
  i+=1
print(b)
```

```python
f=np.zeros((k+1,k+1))
for i in range(1,k):
  for j in range(1,k):
    f[i][j]=-2*betta+alpha*betta*math.pi*(B[j]-B[j]**2)/2*math.cos(math.pi/2*A[i])-2*alpha*betta*math.sin(math.pi/2*A[i])
print(f)
```

```python
u=np.zeros((k+1,k+1))
i=0
for x in B:
  u[0][i]=2*betta*x*(1-x)+2*betta
  i+=1
for j in range(k):
  u[j+1]=np.dot(G,u[j])+b[j+1]+tau*f[j]
print(u)
```

```python
print('t=1',u[k])
print('t=0.5',u[10])
```

```python
c=np.zeros(k+1)
plt.figure(figsize=(8, 8))
x = np.arange(0, 1+h, h)
plt.plot(x, c, linewidth=2, color='red', label="Аналит. решение")
plt.plot(x, u[k], linewidth=2, color='darkorchid', label="Явная схема t=1")

plt.xlabel('x')
plt.grid()
plt.legend()
plt.show()
```

```python
D=1
p=-D*tau/h**2
q=1+2*D*tau/h**2
r=-D*tau/h**2
H=np.zeros((21, 21))
H[0][0]=1
H[k][k]=1
for i in range(1,k):
  H[i][i-1]=p
  H[i][i]=q
  H[i][i+1]=r
print(pd.DataFrame(H))
```

```python
L=np.zeros(k)
M=np.zeros(k)
L[0]=-H[0][1]/H[0][0]
#M[0]=v[1]/H[0][0]

for i in range(1,k):
  L[i]=-H[i][i+1]/(L[i-1]*H[i][i-1]+H[i][i])
#  M[i]=(v[i]-M[i-1]*H[i][i-1])/(L[i-1]*H[i][i-1]+H[i][i])
#M[k]=(v[k]-M[k-1]*H[k][k-1])/(L[k-1]*H[k][k-1]+H[k][k])
```

```python
H_1=np.linalg.inv(H)
```

```python
u1=np.zeros((k+1,k+1))
i=0
for x in B:
  u1[0][i]=2*betta*x*(1-x)+2*betta
  i+=1
for j in range(1,k+1):
  u1[j]=np.dot(H_1,u1[j-1]-b[j-1]+b[j]+tau*f[j])
print(u1)
```
